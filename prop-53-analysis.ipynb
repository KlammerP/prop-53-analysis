# Prop 53 analysis
This is my first own python data analysis, run on data from CalAccess provided by the California Civic Data Coalition. I want to analyze the major donors to November 2016 proposition 53 ballot measure committees.

```python
import pandas as pd
```

Read in proposition data from CSV

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head()
```

```python
props.info()
```

Read in contributors' (donors') data from CSV

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

```python
contribs.head()
```

```python
props.prop_name.value_counts()
```

Single out only committees concerning prop 53

```python
prop53 = props[props.prop_name == "PROPOSITION 053 - REVENUE BONDS. STATEWIDE VOTER APPROVAL. INITIATIVE CONSTITUTIONAL AMENDMENT."]
```

```python
prop53.head()
```

```python
prop53.info()
```

```python
contribs.info()
```

Build a merged data table that only includes contributions to pro 53 committees

```python
merged = pd.merge(prop53, contribs, on="calaccess_committee_id")
```

```python
merged.info()
```

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum() * 100
```

The highest single contributions for prop 53 committees

```python
merged.sort_values("amount", ascending=False)
```

```python
oppose.sort_values("amount", ascending=False).head()
```

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

The biggest donors by last name sorted by amount

```python
merged.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
